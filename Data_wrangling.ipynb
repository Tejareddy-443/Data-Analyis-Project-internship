```python
df=pd.read_csv("Cleaned_Sample_Superstore.csv",encoding='latin1')
df.head()
```

```python
print(df["Category"].value_counts())
print(df["Region"].value_counts())
```

```python
df["Sales"].describe()
```

“The sales distribution is positively skewed, indicating that the majority of transactions generate lower sales values, while only a few transactions contribute exceptionally high revenue.”

```python
import matplotlib.pyplot as plt
plt.hist(df['Sales'],bins=50)
plt.xlim(0,5000)
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

“Technology products generated the highest total sales among all categories, indicating strong customer demand and revenue contribution. Furniture and Office Supplies also contributed significantly but at comparatively lower levels.”

```python
category_sales=df.groupby("Category")["Sales"].sum()
print()
category_sales.plot(kind="bar",color=['blue', 'skyblue', 'violet'], figsize=(10,5))
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

```python
shipmode_frequency=df.groupby("Ship Mode")["Quantity"].sum()
print(shipmode_frequency)
shipmode_frequency.plot(kind="bar",color=['blue', 'green', 'orange', 'purple'], figsize=(8,4))
plt.title("Quantity by Ship Mode")
plt.xlabel("Ship Mode")
plt.ylabel("Total Quantity")
plt.show()
```

```python
# Convert Order Date to datetime
df["Order Date"] = pd.to_datetime(df["Order Date"])

# Extract Month and Day
df["Month"] = df["Order Date"].dt.month
df["Day"] = df["Order Date"].dt.day
```

```python
monthly_profit=df.groupby("Month")["Profit"].sum()
monthly_profit.plot(kind="line", marker='o', color='blue', figsize=(10,5))
plt.title("Monthly Profit Trend")
plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.show()
```
